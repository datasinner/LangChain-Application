```python
from langchain import LLMChain, PromptTemplate
from langchain.llms import OpenAI

# Azure OpenAI credentials
api_key = "4c81e9b0eae04972a9132ea399d6746a"
endpoint = "https://ai-proxy.lab.epam.com/"
deployment_name = "Test_1"

# Initialize the AzureOpenAI model in LangChain
llm = OpenAI(
    openai_api_key=api_key,
    openai_api_base=endpoint,
    openai_api_type="azure",
    openai_api_version="2023-05-15",
    deployment_name=deployment_name,
)

# Create a prompt template
prompt = PromptTemplate.from_template(
    "Translate the following English text to French: {text}"
)

# Create a LangChain pipeline
chain = LLMChain(llm=llm, prompt=prompt)

# Use the chain to generate output
response = chain.run(text="Hello, how are you?")
print(response)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.chains import RunnableSequence

# Azure OpenAI credentials
api_key = "4c81e9b0eae04972a9132ea399d6746a"
endpoint = "https://ai-proxy.lab.epam.com/"
deployment_name = "Test_2"

# Initialize the AzureOpenAI model in LangChain
llm = ChatOpenAI(
    openai_api_key=api_key,
    openai_api_base=endpoint,
    openai_api_type="azure",
    openai_api_version="2023-05-15",
    model=deployment_name,
)

# Create a prompt template
prompt = PromptTemplate.from_template(
    "Translate the following English text to French: {text}"
)

# Create a RunnableSequence (replacing LLMChain)
sequence = RunnableSequence([prompt, llm])

# Use the sequence to generate output
response = sequence.invoke({"text": "Hello, how are you?"})
print(response)
```
